```
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```
f0 = 60
L = 2
b = [1, 0, (2.0*np.pi*f0)**2]
a = [1, 2*np.pi*L, (2.0*np.pi*f0)**2]

w, H = signal.freqs(b,a,worN=np.arange(0,2*np.pi*120,0.01))

plt.plot(w/(2*np.pi), 20*np.log10(abs(H)))
plt.title('Resposta em magnitude de um filtro Notch analógico')
plt.xlabel('Frequencia [Hz]')
plt.ylabel('Resposta em Magnitude [dB]')
plt.grid('minor')
plt.axvline(59, color='green')
plt.axvline(61, color='green')
plt.axhline(-3, color='green')
```

```
plt.plot(w/(2*np.pi), 20*np.log10(abs(H)))
plt.title('Resposta em magnitude de um filtro Notch analógico')
plt.xlabel('Frequencia [Hz]')
plt.ylabel('Resposta em Magnitude [dB]')
plt.grid('minor')
plt.axvline(59, color='green')
plt.axvline(61, color='green')
plt.axhline(-3, color='green')
plt.axis([0.9*(f0-L), 1.1*(f0+L), -4, 0])
```

```
fs = 240

omega0 = 2.0*np.pi*f0/fs
B = 2*np.pi*L/fs

alpha = (1.0-np.tan(B/2))/(1.0+np.tan(B/2))
print(alpha)
beta = np.cos(omega0)
print(beta)

bdigital = [0.5*(1+alpha), -beta*(1+alpha),0.5*(1+alpha)]
adigital = [1, -beta*(1+alpha), alpha]


w, H = signal.freqz(bdigital, adigital)

plt.plot(fs*w/(2*np.pi), 20*np.log10(abs(H)))
plt.title('Resposta em magnitude de um filtro Notch digital')
plt.xlabel('Frequencia [Hz]')
plt.ylabel('Resposta em Magnitude [dB]')
plt.grid('minor')
plt.axvline(59, color='green')
plt.axvline(61, color='green')
plt.axhline(-3, color='green')
plt.axis([0, fs/2, -80, 5])
```
